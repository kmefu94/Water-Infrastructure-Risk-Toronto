This module seeks to split the complex dataset of ward-profiles, which contains several workbooks within the excel file.
Additionally, due to the vast amount of census data collected, each category would need to be split into it's own dataset as well.

```python
import sys
from pathlib import Path
import pandas as pd

# add src to path
sys.path.append(str(Path.cwd().parents[1] / "src"))

from utils.data_loader import pull_data
from utils.data_loader import read_files
```

```python
ROOT = Path.cwd().parents[1]   # project/
DATA_PATH = ROOT / "data/RawData/ward-profiles-25-ward-model"
```

```python
EXCLUDED_WARD_FILES = {
   "2023-wardprofiles-geographicareas.xlsx",
}
```

```python
def load_target_workbook_sheets(folder_path, target_stem, excluded_files=None):
    folder_path = Path(folder_path)
    excluded_files = excluded_files or set()

    for file_path in folder_path.glob("*.xls*"):
        if file_path.name in excluded_files:
            continue

        if file_path.stem.lower() == target_stem.lower():
            xls = pd.ExcelFile(file_path)

            return {
                sheet.lower().replace(" ", "_"): pd.read_excel(file_path, sheet_name=sheet)
                for sheet in xls.sheet_names
            }

    raise FileNotFoundError(f"{target_stem} not found in {folder_path}")
```

```python
ward_profile_sheets = load_target_workbook_sheets(
    DATA_PATH,
    "2023-wardprofiles-2011-2021-censusdata_rev0719_modified",
    excluded_files=EXCLUDED_WARD_FILES
)
```
